```python
from lxml import html
import requests
import pandas as pd
```

```python
header = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 YaBrowser/23.7.4.971 Yowser/2.5 Safari/537.36'}
url = 'https://lenta.ru/'
response = requests.get(url=url, headers=header)
dom = html.fromstring(response.text)
```

```python
news_info = []
```

```python
title_news = {}
```

```python
t_title = dom.xpath("//h3[contains(@class, 'card-big__title')]/text()")
t_link = dom.xpath("//a[contains(@class, '_topnews')]/@href")
t_date = dom.xpath("//a[contains(@class, '_topnews')]/@href")
```

```python
t_source = url
t_title = t_title[0]
t_link = url[0:16]+t_link[0]
t_date = (t_date[0]).split(sep='/')
t_date = '.'.join(t_date[2:5])
```

```python
title_news['source'] = t_source
title_news['title'] = t_title
title_news['link'] = t_link
title_news['date'] = t_date
```

```python
t_news = []
t_news.append(title_news)
t_news = pd.DataFrame(t_news)
t_news
```

```python
ot_news = []
```

```python
titles = dom.xpath("//a[contains(@class, '_topnews')]//h3[contains(@class, 'card-mini__title')]/text()")
```

```python
for title in titles:
    news_others = {}
    news_others['source'] = url
    news_others['title'] = title
    ot_news.append(news_others)
```

```python
ot_news = pd.DataFrame(ot_news)
ot_news
```

```python
ot_news_links = []
```

```python
links = dom.xpath("//a[contains(@class, '_topnews')]/@href")

i = 0
for link in links:
    news_others_links = {}
    i += 1
    date = link.split(sep='/')
    date = '.'.join(date[2:5])
    link = url[0:16] + link
    if i > 1:
        news_others_links['link'] = link
        news_others_links['date'] = date
        ot_news_links.append(news_others_links)
    else:
        pass
    
```

```python
ot_news_links = pd.DataFrame(ot_news_links)
ot_news_links
```

```python
news = pd.concat([ot_news, ot_news_links], axis = 1)
news
```

```python
all_news = pd.concat([t_news, news])
```
